# 4. Apply the MAX - MIN - AVERAGE functions to raisedAmt column

```python
#df.describe()

maxValue = df['raisedAmt'].max()
print("Maximum value in column 'raisedAmt': " , maxValue)

minValue = df['raisedAmt'].min()
print("Minimum value in column 'raisedAmt': " , minValue)

meanValue = df['raisedAmt'].mean()
print("Average value in column 'raisedAmt': " , meanValue)
```
